```python
df = pd.DataFrame(documents)
df.columns = ["Título", "Conteúdo"]
df
```

```python
def embed_fn(title, text):
  return genai.embed_content(model=model,
                                content=sample_text,
                                task_type="retrieval_document",
                                title=title)["embedding"]
```

```python
df["Embaddings"] = df.apply(lambda row : embed_fn(row["Título"], row["Conteúdo"]), axis=1)
df
```

```python
def gerar_buscar_consulta(consulta, base, model):
  embedding_consulta = genai.embed_content(model=model,
                                content=consulta,
                                task_type="RETRIEVAL_QUERY")["embedding"]

  produtos_escalares = np.dot(np.stack(df["Embaddings"]), embedding_consulta)

  indice = np.argmax(produtos_escalares)

  return df.iloc[indice]["Conteúdo"]
```

```python
consulta = "como funciona a marcha no carro do google"
gerar_buscar_consulta(consulta, df, model)
```
